```python
from gnews import GNews
google_news = GNews(start_date=(2023,1,1), end_date=(2023,9,1))
results = google_news.get_news(' Unified Payment Interface (UPI)')
```

```python
import pandas as pd
from newspaper import Article


df = pd.DataFrame(results)

# Extract only the 'title' from the 'publisher' column
df['source'] = df['publisher'].apply(lambda x: x['title'])
df['news_website_url'] = df['publisher'].apply(lambda x: x['href'])
df.drop('publisher',axis=1,inplace=True)
def get_full_article(row):
    try:
        article = Article(url=row["url"])
        article.download()
        article.parse()
        return article.text
    except Exception as e:
        return pd.NA  # Set it to NaN if an error occurs
df["full_article"] = df.apply(get_full_article,axis=1)
# Specify the CSV file path
csv_file = "news_articles.csv"

# Write the DataFrame to a CSV file
df.to_csv(csv_file, index=False)

print(f'Data has been written to {csv_file}.')
```
