```python
import pandas as pd

# Load and display the first few rows of each dataset to understand its structure
dataset1 = pd.read_excel('C:\\Users\\PC2\\Downloads\\Ver04\\TFND_2019_Ver01.xlsx')
dataset2 = pd.read_excel('C:\\Users\\PC2\\Downloads\\output.xlsx')


```

```python
# Merge the datasets on 'dID' column
merged_data = pd.merge(dataset1, dataset2, on='dID', how='inner')

# Selecting relevant features and the label
features_and_label = merged_data[[
    'Zem_WordCount','MainSource',
     'Orj_Text_sentiment','Orj_WordCount', 'Class'
]]

# Display the first few rows of the merged and selected data
features_and_label.head()
```

```python
features_and_label.isna().sum()
```

```python
features_and_label.dropna(inplace=True)
features_and_label.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le_source = LabelEncoder()
features_and_label['MainSource'] = le_source.fit_transform(features_and_label['MainSource'])

# Encoding categorical data
le_sentiment = LabelEncoder()
features_and_label['Orj_Text_sentiment_encoded'] = le_sentiment.fit_transform(features_and_label['Orj_Text_sentiment'])

# Encoding the Class label
le_class = LabelEncoder()
# change true to True inn (features_and_label['Class']
features_and_label['Class'] = features_and_label['Class'].apply(lambda x: 'TRUE' if x == True else x)
features_and_label['Class_encoded'] = le_class.fit_transform(features_and_label['Class'])


```

```python
# remove class_encoded 2 and 4 from the feature_label
features_and_label = features_and_label[features_and_label['Class_encoded'] != 2]
features_and_label = features_and_label[features_and_label['Class_encoded'] != 4]

features_and_label['Class_encoded'].value_counts()
```

```python
final_features = features_and_label.drop(columns=['Orj_Text_sentiment', 'Class'])
final_label = features_and_label['Class_encoded']
final_features.head()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, train_test_split

# Initialize the Support Vector Classifier
svc = SVC(random_state=42)

# Define the parameter grid for hyperparameter tuning of SVM
param_grid = {
    'C': [0.1, 1, 10, 100],  # Regularization parameter
}

# Remove the 'Class_encoded' feature from the feature set
features_final = final_features.drop(columns=['Class_encoded'])

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    features_final, final_label, test_size=0.2, random_state=42
)

# Set up the grid search with cross-validation for the SVM model
grid_search = GridSearchCV(svc, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Get the best parameters, best cross-validation score, and evaluate on the test set
best_params = grid_search.best_params_
best_score = grid_search.best_score_
best_model = grid_search.best_estimator_
test_accuracy = best_model.score(X_test, y_test)

(best_params, best_score, test_accuracy)
```
