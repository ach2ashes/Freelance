```python
def project_num(input_data):
    n=0
    for i in  input_data.keys():
        if not isinstance(i,tuple) : 
            n+=1
    return n
```

```python
from itertools import combinations

input_data = {
    'A': 20,
    'B': 20,
    'C': 20,
    'D': 20,
    'E': 20,
    'F': 20,
    'G': 20,
    'H': 20,
    'I': 20,
    'J': 20,
    ('G', 'I'): 18,
    ('H', 'I'): 17,
    ('D', 'E'): 19,
    ('A', 'B'): 16,
    ('B', 'C'): 17,
    ('B', 'F'): 15,
    ('B', 'H'): 18,
    ('C', 'F'): 16,
    ('C', 'H'): 14,
    ('F', 'H'): 17,
    ('G', 'H', 'I'): 13,
    ('A', 'B', 'C'): 12,
    ('F', 'G', 'H'): 15,
    ('B', 'F', 'H'): 12,
    ('B', 'C', 'F'): 11,
    ('B', 'C', 'H'): 13,
    ('C', 'F', 'H'): 10,
    ('B', 'C', 'F', 'H'): 8,
}

# Helper function to flatten nested tuples
def flatten_tuple(t):
    flattened = []
    for item in t:
        if isinstance(item, tuple):
            flattened.extend(flatten_tuple(item))
        else:
            flattened.append(item)
    return flattened

# Helper function to calculate the price of a combination
def calculate_price(combination):
    price = 0
    for item in combination:
            price += input_data[item]
    return price

# Helper function to check if a combination contains duplicate items
def has_duplicates(combination):
    comb = combination
    seen = set()
    flattened_combo = flatten_tuple(comb)
    if  len(flattened_combo) ==len(set(flattened_combo)):
         return True
    return False
def helpfilt(t):
  l = 0
  for i in t :
    if isinstance(i,tuple):
      l +=len(i)
    else :
      l+=1
  return l

# Define a generator function to yield combinations
def generate_combinations():
    for r in range(2, len(input_data) + 1):
        for combo in combinations(input_data.keys(), r):
            yield combo
l ={}
# Print the desired combinations using the generator
for combo in generate_combinations():
    if isinstance(combo, tuple) and helpfilt(combo) == project_num(input_data) and has_duplicates(combo):
        price = calculate_price(combo)
        l[combo]=price
```

```python
sorted_combo_price_dict = {k: v for k, v in sorted(l.items(), key=lambda item: item[1])}
import csv

# Assuming you have sorted_combo_price_dict or sorted_combo_price_dict_desc from the previous step

# Specify the CSV file name
csv_file_name = "sorted_combos.csv"

# Open the CSV file in write mode
with open(csv_file_name, mode='w', newline='') as csv_file:
    # Create a CSV writer
    csv_writer = csv.writer(csv_file)
    
    # Write the header row (optional)
    csv_writer.writerow(["Combo", "Price"])  # Replace with your column headers
    
    # Write the data from the sorted dictionary
    for combo, price in sorted_combo_price_dict.items():
        csv_writer.writerow([combo, price])

# Now, the sorted data has been written to the CSV file "sorted_combos.csv"
```
