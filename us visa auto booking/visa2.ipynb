```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.service import Service
from datetime import datetime, timedelta


option = webdriver.ChromeOptions()
service = Service(executable_path="C:\\Users\\PC2\\Downloads\\chromedriver-win64 (3)\\chromedriver-win64\\chromedriver.exe")
#option.add_argument("--headless")
#option.add_argument("user-data-dir=C:\\Users\\PC2\\AppData\\Local\\Google\\Chrome\\User Data")#to open a window with your default profile (aka your google acc)
driver=webdriver.Chrome(service = service,options=option)
```

```python
import time       

city = "Monterrey"    
def start(email,password,city)   :                        #importing the time module       
    driver.get("https://ais.usvisa-info.com/es-mx/niv/users/sign_in")
    driver.implicitly_wait(10)
    driver.find_element(By.XPATH, "//input[@name='user[email]']").send_keys(email) 
    driver.find_element(By.XPATH, "//input[@name='user[password]']").send_keys(password) 
    ok = driver.find_element(By.XPATH, "//input[@name='policy_confirmed']")
    driver.execute_script("arguments[0].click();", ok)
    driver.find_element(By.XPATH, "//input[@name='commit']").click()
    driver.find_element(By.XPATH,"//a[contains(text(),'Continuar')]").click()
    driver.find_element(By.XPATH,"//span[@class='fas fa-calendar-alt fa-lg fa-fw']").click()
    driver.find_element(By.XPATH,"//a[contains(text(),'Programe la cita')]").click()
    dropdown = Select(driver.find_element(By.XPATH, "//select[@name='appointments[consulate_appointment][facility_id]']"))
    dropdown.select_by_visible_text(city)
```
